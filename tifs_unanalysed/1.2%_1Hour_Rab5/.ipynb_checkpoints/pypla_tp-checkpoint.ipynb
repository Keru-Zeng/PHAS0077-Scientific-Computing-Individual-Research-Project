```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
mpl.rc('figure',  figsize=(10, 6))
import numpy as np
import pandas as pd
from pandas import DataFrame, Series  # for convenience
from scipy.spatial import distance as dist
import scipy.cluster.hierarchy as hier

import sys
sys.path.insert(0, "/Users/joe/Modules")
from pypla import particle
from pypla import preprocessor

import pims
import trackpy as tp

tifdir = './tifs'
files, metadata = preprocessor.data_org(tifdir)

columns = ('x', 'y', 'z', 'a_total', 'intensity', 'particle', 'com_dist', 'trial')
all_parts = pd.DataFrame([], columns=columns)
upper_mass_cut = 10000

for i in range(0,len(files)):
    print(i)
    print(files[i])
    frames = pims.TiffStack('./tifs/' + files[i])
    nps = np.array(frames[::2])
    nuclei = np.array(frames[1::2])
    features = tp.batch(nps[:], diameter=(5, 5))
    features = features[features['raw_mass'] > 0]

    if len(features) == 0:
        features['particle'] = []
    if len(features) == 1:
        features['particle'] = 1
    elif len(features) > 1:
        # Clustering to eliminate maxima that are too close together
        positions = features[['x', 'y', 'frame']].values
        # Distance matrix is n-particles x n-particles in size - reckon it gives the interparticle separation
        # This gives the upper triangle of the distance matrix 
        dist_mat = dist.pdist(positions)
        link_mat = hier.linkage(dist_mat)
        # fcluster assigns each of the particles in positions a cluster to which it belongs
        cluster_idx = hier.fcluster(link_mat, 5, criterion='distance')
        features['particle'] = cluster_idx
    
    n_parts = np.unique(features['particle'].values)
    values_clustered = []
    for j in n_parts:
        current = features[features['particle'] == j]
        if len(current) > 1:
            norm = np.sum(current['raw_mass'])
            x_av = np.mean(current['x'])
            y_av = np.mean(current['y'])
            #z-value is the intensity-weighted value
            z_av = np.sum(current['raw_mass'] * current['frame'])/norm
            a_total = np.sum(current['size'])
                # Geometric cut-off to particle size
            if a_total > 1:
                values_clustered.append([x_av, y_av, z_av, a_total, norm, j])
    
    columns_clustered = ('x', 'y', 'z', 'a_total', 'intensity', 'particle')
    particles_clustered = pd.DataFrame(values_clustered, columns=columns_clustered)
    particles_clustered = particles_clustered[particles_clustered['raw_mass'] < 10000]
    print('Found', len(particles_clustered), 'particles')

    plt.clf()
    fig = plt.figure()
    ax = fig.add_subplot(111)
#     ax.imshow(np.mean(frames, axis=0), cmap='gray', vmin=0, vmax=255)
    ax.imshow(nps.max(axis=0), cmap='gray', vmin=0, vmax=255)
    ax.scatter(particles_clustered['x'], particles_clustered['y'], s=50, facecolor='none', edgecolor='red')
    plt.savefig(files[i] + '_particles.tif', bbox_inches='tight')
    
    # Numbers purely for visualisation purposes
    plt.clf()
    xy_scale, z_scale = 1., 2.
    # Plot 3D visualisation of data
    fig = plt.figure(figsize=(12, 12))
    # xy projection:
    ax_xy = fig.add_subplot(111)
    ax_xy.imshow(nps.max(axis=0), cmap='gray')
    ax_xy.scatter(particles_clustered['x'], particles_clustered['y'], facecolors='none', edgecolors='red', s=100)

    divider = make_axes_locatable(ax_xy)
    ax_zx = divider.append_axes("top", 2, pad=0.2, sharex=ax_xy)
    ax_zx.imshow(nps.max(axis=1), aspect=z_scale/xy_scale, cmap='gray')
    ax_zx.scatter(particles_clustered['x'], particles_clustered['z'], facecolors='none', edgecolors='red', s=100)
    ax_yz = divider.append_axes("right", 2, pad=0.2, sharey=ax_xy)
    ax_yz.imshow(nps.max(axis=2).T, aspect=xy_scale/z_scale, cmap='gray')
    ax_yz.scatter(particles_clustered['z'], particles_clustered['y'], facecolors='none', edgecolors='red', s=100)
    plt.savefig(files[i] + '_particles.tif', bbox_inches='tight')
    
    particles_clustered.to_csv(files[i] + '.csv')
    
    #Calculate the z-profile of the nucleus image
    profile = np.mean(nuclei, axis=1)
    z_profile = np.mean(profile, axis=1)
    #Locate maximum in z-profile
    z0 = np.where(z_profile == z_profile.max())[0]
    particles_clustered['com_dist'] = particles_clustered['z'] - float(z0)
    particles_clustered['trial'] = i
    all_parts = all_parts.append(particles_clustered)
    
    
```
